```
from icrawler.builtin import GoogleImageCrawler
animal_names = ['パンダ', 'ゾウ', 'ホッキョクグマ', 'ゴリラ', 'トラ', 'ペンギン', 'フラミンゴ', 'ハシビロコウ', 'サル', 'キリン', 'サイ', 'アシカ', 'カワウソ', 'カバ',  'ワニ',  'ライオン', 'カンガルー', 'シマウマ', 'アザラシ', 'レッサーパンダ']
filters = dict(size='large', type="photo")
for name in animal_names:
    crawler = GoogleImageCrawler(storage = {"root_dir": "drive/My Drive/kikagaku/longterm/UenoAnimalClassification/images"+"/"+name})
    crawler.crawl(keyword=name, max_num = 100, filters = filters)
```

```
import os
import glob

# ふぃいる名をいったん変更する
for name in animal_names:
    path = "drive/My Drive/kikagaku/longterm/UenoAnimalClassification/images/" + name
    files = glob.glob(path+"/*")
    
    for i, f in enumerate(files, 1):
        os.rename(f, os.path.join(path, '{0:03d}'.format(i) + '_google_' + os.path.basename(f)))
```

```
from icrawler.builtin import BaiduImageCrawler, BingImageCrawler, GoogleImageCrawler

animal_names = ['パンダ', 'ゾウ', 'ホッキョクグマ', 'ゴリラ', 'トラ', 'ペンギン', 'フラミンゴ', 'ハシビロコウ', 'サル', 'キリン', 'サイ', 'アシカ', 'カワウソ', 'カバ',  'ワニ',  'ライオン', 'カンガルー', 'シマウマ', 'アザラシ', 'レッサーパンダ']

for name in animal_names:
    bing_crawler = BingImageCrawler(storage = {"root_dir": "drive/My Drive/kikagaku/longterm/UenoAnimalClassification/images"+"/"+name})
    filters = dict(size='large', type="photo")
    bing_crawler.crawl(keyword=name, max_num = 100, filters=filters)
```
